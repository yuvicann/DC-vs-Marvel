# LINEAR REGRESSION MODEL ON THE MARVEL DATASET
TO DEPICT THE IMDB RATINGS BASED ON THE META SCORE OF THE MOVIE

```python
new_data = marvel_data[['Original Title','Company','Rate','Metascore','Minutes','Release','Budget','Gross Worldwide']]
new_data.head(5)
```

```python
sns.scatterplot(x=new_data['Metascore'],y=new_data['Rate'])
plt.xlabel('Metascore of MCU Movies')
plt.ylabel('IMDB Ratings')
plt.title('Metascore vs IMDB Ratings of MCU Movies')
```

```python
#train:test = 70:30
mask = np.random.rand(len(new_data))<0.70
train = new_data[mask]
test = new_data[~mask]
```

```python
sns.scatterplot(x=train['Metascore'],y=train['Rate'])
plt.xlabel('Metascore of MCU Movies')
plt.ylabel('IMDB Ratings')
plt.title('Training set')
```

```python
from sklearn import linear_model #finding the slope and intercept of the fit line
regr = linear_model.LinearRegression()
train_x = np.asanyarray(train[['Metascore']])
train_y = np.asanyarray(train[['Rate']])
regr.fit (train_x, train_y)
# The coefficients
print ('Coefficients: ', regr.coef_)
print ('Intercept: ',regr.intercept_)
```

```python
plt.scatter(train['Metascore'], train['Rate'],  color='red', alpha = 0.7)
plt.plot(train_x, regr.coef_[0][0]*train_x + regr.intercept_[0], '-r', color='blue') #y=mx+c
plt.xlabel("IMDB Ratings")
plt.ylabel("Worldwide collection of movies")
plt.title("Fit Line over Scatter plot of Train Set")
```

```python
from sklearn.metrics import r2_score #predicting test results and printing out the error

test_x = np.asanyarray(test[['Metascore']])
test_y = np.asanyarray(test[['Rate']])
test_y_ = regr.predict(test_x)
mae = np.mean(np.absolute(test_y_ - test_y))
mse = np.mean((test_y_ - test_y)**2)
r2 = r2_score(test_y_ , test_y)
print("Mean absolute error:", mae)
print("Residual sum of squares (MSE):", mse)
print("R2-score:", r2 )
```

# Predict the imdb rating when metascore is 69.

```python
print("The IMDB Ratings when Metascore is 69 will be:",regr.predict([[69]]))
```

# LINEAR REGRESSION ON THE DC DATASET
TO DEPICT THE IMDB RATINGS BASED ON THE META SCORE OF THE MOVIE

```python
new_data = DC_data[['Original Title','Company','Rate','Metascore','Minutes','Release','Budget','Gross Worldwide']]
new_data.head(5)
```

```python
sns.scatterplot(x=new_data['Metascore'],y=new_data['Rate'])
plt.xlabel('Metascore')
plt.ylabel('IMDB Ratings')
plt.title('IMDB Ratings w.r.t Metascore of The Movies')
```

```python
#train:test = 70:30
mask = np.random.rand(len(new_data))<0.70
train = new_data[mask]
test = new_data[~mask]
```

```python
sns.scatterplot(x=train['Metascore'],y=train['Rate'])
plt.xlabel('Metascore')
plt.ylabel('IMDB Ratings')
plt.title('Training set')
```

```python
from sklearn.metrics import r2_score #predicting test results and printing out the error

test_x = np.asanyarray(test[['Metascore']])
test_y = np.asanyarray(test[['Rate']])
test_y_ = regr.predict(test_x)
mae = np.mean(np.absolute(test_y_ - test_y))
mse = np.mean((test_y_ - test_y)**2)
r2 = r2_score(test_y_ , test_y)
print("Mean absolute error:", mae)
print("Residual sum of squares (MSE):", mse)
print("R2-score:", r2 )
```
